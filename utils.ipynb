```python
import json

def policies_to_json(file_path, json_path):
    reading_policies = False
    policy_dict = {}
    
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            if line == "[policies]":
                reading_policies = True
                continue
            if reading_policies and '=' in line:
                key, value = line.split('=')
                policy_dict[key.strip()] = float(value.strip())

    with open(json_path, 'w') as jf:
        json.dump(policy_dict, jf, indent=4)

policies_to_json("simulation/countries/france/france.txt", "current_french_policies.json")
```
